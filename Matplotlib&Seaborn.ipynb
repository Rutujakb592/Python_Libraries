```python
import pandas as pd
data = pd.read_csv("vgsales.csv")
data.head()
```

```python
data['Genre']
```

```python
data['Genre'].unique()
```

```python
count = data['Genre'].value_counts()
count
```

```python
x_bar = count.index
y_bar = count.values
plt.bar(x_bar, y_bar)
```

```python
plt.bar(x_bar, y_bar, width=0.2, color='orange')
plt.xticks(rotation=90, fontsize=10)


plt.xlabel('Genre', fontsize=12)
plt.ylabel('count', fontsize=12)
plt.title('Top Genre Analysis', fontsize=18)
plt.show()
```

```python
sns.countplot(x='Genre', data=data, order=data['Genre'].value_counts().index, color='orange')
```

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_data
```

```python
region_sales = sales_data.T.sum(axis='columns')
region_sales
```

```python
plt.pie(region_sales,
        labels=region_sales.index,
        startangle=90,
        explode=(0.2, 0, 0, 0),
        autopct='%1.1f%%')
plt.title('Total sales across various regions')

plt.show()
```

1) plot() - Connects points with line, Used for trends
2) scatter() - Shows only points, Used for relationships

```python
#top 3 platform
plt.figure(figsize=(12, 8))
sns.countplot(x='Publisher', data=data, hue='Platform')
```

```python
#top 5 publisher
top_3_pub = data['Publisher'].value_counts().index[:3]
top_3_pub
```

```python
top_5_Genre =data['Genre'].value_counts().index[:3]
top_5_Genre
```

```python
top_5_plat =data['Platform'].value_counts().index[:3]
top_5_plat
```

```python
top_5_data = data.loc[(data['Publisher'].isin(top_3_pub) & data['Platform'].isin(top_5_plat) & data['Genre'].isin(top_5_Genre))]
top_5_data
```

```python
#Dodge Count Plot
sns.countplot(data=top_5_data, x='Publisher', hue='Platform')
```

```python
#Stacked plot
df_stacked_plt = pd.crosstab(index=top_5_data['Publisher'], columns=top_5_data['Platform'])
df_stacked_plt
```

```python
df_stacked_plt.plot(kind='bar', stacked=True)
plt.show()
```

```python
sns.boxplot(x='Publisher', y='Global_Sales', data=top_5_data)
```
